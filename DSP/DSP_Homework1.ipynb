```python
from thinkdsp import CosSignal, SinSignal

signal1 = CosSignal(freq=261.626, amp=1.0, offset=0)
signal2 = CosSignal(freq=294.33, amp=1.0, offset=0)
signal3 = CosSignal(freq=327.03, amp=1.0, offset=0)
signal4 = CosSignal(freq=348.83, amp=1.0, offset=0)
signal5 = CosSignal(freq=392.44, amp=1.0, offset=0)
signal6 = CosSignal(freq=436.04, amp=1.0, offset=0)
signal7 = CosSignal(freq=490.55, amp=1.0, offset=0)
signal8 = CosSignal(freq=523.25, amp=1.0, offset=0)


do = signal1.make_wave(duration=1, start=0, framerate=11025)
re = signal2.make_wave(duration=1, start=1, framerate=11025)
mi = signal3.make_wave(duration=1, start=2, framerate=11025)
fa = signal4.make_wave(duration=1, start=3, framerate=11025)
sol = signal5.make_wave(duration=1, start=4, framerate=11025)
la = signal6.make_wave(duration=1, start=5, framerate=11025)
si = signal7.make_wave(duration=1, start=6, framerate=11025)
do1 = signal8.make_wave(duration=2, start=7, framerate=11025)

wave = do+re+mi+fa+sol+la+si+do1

#wave = do.make_wave(duration=5, start=0, framerate=11025)
wave.normalize()
wave.apodize()
wave.make_audio()
```
