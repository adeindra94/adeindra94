```python
#Gamma
a = 1.99
mean, var, skew, kurt = st.gamma.stats(a, moments='mvsk')
x = np.linspace(st.gamma.ppf(0.01,a),
                 st.gamma.ppf(0.99,a), 100)
print(mean,var,skew,kurt)
print(x)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, st.gamma.pdf(x,a),
       'r-', lw=5, alpha=0.6, label='Gamma PDF')
rv = st.gamma(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='Frozen PDF')
vals = st.gamma.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], st.gamma.cdf(vals, a))
r = st.gamma.rvs(a, size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```
