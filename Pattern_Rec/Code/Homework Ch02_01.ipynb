```python
import sympy as sm
from sympy import *
x, y, z=symbols('x y z')
init_printing(use_unicode=True)
```

```python
x, y, z = sm.symbols('x,y,z')
x, y, z
```

```python
f = 3*x**7 + 5*x**3 - 11
f
```

```python
devir = diff(f)
devir
```

```python
integ = integrate(f)
integ
```

```python
from sympy.plotting import plot
```

```python
plot(f)
```
